```python
import read_ucr
import os
import pandas as pd
import numpy as np

PATH_TO_DATA       = os.path.join(".", "ucr_datasets")
```

```python
VAR_LENGTH_DATASETS = ["CharacterTrajectories", "SpokenArabicDigits", "JapaneseVowels"]
FIX_LENGTH_DATASETS = ["ArticularyWordRecognition", "ERing", "Plane", "Cricket", "Mallat", "UWaveGestureLibrary", "Symbols", "PenDigits", "Fungi", "NATOPS", "ECG5000"]
```

```python
def znormalize(series):
    series = (series - np.mean(series, axis=None)) / np.std(series, axis=None)
    return series
```

```python
# Concatenate test and train, z-normalize, and store 
for ds_name in FIX_LENGTH_DATASETS + VAR_LENGTH_DATASETS:
    print(ds_name)
    path_to_dataset = os.path.join(PATH_TO_DATA, ds_name.lower())
    
    df_train, df_test = read_ucr.load_ucr_dataset(PATH_TO_DATA, ds_name, train=True), read_ucr.load_ucr_dataset(PATH_TO_DATA, ds_name, train=False)
    df = pd.concat((df_train, df_test), axis=0).reset_index(drop=True)
    
    # CharacterTrajectories is already normalized
    if ds_name != "CharacterTrajectories":
        df['series'] = df['series'].apply(znormalize)
        
    path_to_df = os.path.join(path_to_dataset, "znormalized.pkl")
    if not os.path.exists(path_to_df):
        df.to_pickle(path_to_df)
```
